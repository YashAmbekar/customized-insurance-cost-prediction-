Machine Learning model Development

```python
insurance_data.head()
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
insurance_data["sex"] = le.fit_transform(insurance_data["sex"])
insurance_data["smoker"] = le.fit_transform(insurance_data["smoker"])
insurance_data["region"] = le.fit_transform(insurance_data["region"])
```

```python
insurance_data.head()
```

Splitting Data into dependent and independent variable 
x = independent variable
y = dependent variable 

```python
x = insurance_data.drop(columns=["expenses"], axis=1)
y = insurance_data["expenses"]
```

```python
x
```

```python
y
```

Splitting the data into training and texting data 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2 ,random_state= 42, shuffle=True)

## checking the shape of splitting data 
print(f"the shape of x trian : {x_train.shape}")
print(f"the shape of y train : {y_train.shape}")

print("-"*30)

print(f"the shape of x test : {X_test.shape}")
print(f"the shape of y test : {y_test.shape}")
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
x["age"] = std_scaler.fit_transform(np.array(x["age"]).reshape(-1,1))
x["bmi"] = std_scaler.fit_transform(np.array(x["bmi"]).reshape(-1,1))
x["children"] = std_scaler.fit_transform(np.array(x["children"]).reshape(-1,1))
```

```python
x # now the data is converted to a somilar scale 
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg_model = LinearRegression()
```

```python
linear_reg_model.fit(x_train, y_train)
```

Model Cross Validation

```python
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
result = cross_validate(linear_reg_model, x, y, cv = 5, return_train_score=True, scoring= "r2")
result
```

```python
train_acc = result["train_score"].mean()*100
print("Training Accuracy = ", round(train_acc, 2), "%")
```

```python
test_acc = result["test_score"].mean() * 100
print("Testing Accuracy = ", round(test_acc, 2), "%")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
models = []
models.append(("LinearRegression", LinearRegression()))
models.append(("Knn", KNeighborsRegressor()))
models.append(("Decision Tree", DecisionTreeRegressor()))
models.append(("Random Forest", RandomForestRegressor()))
for model in models:
    result = cross_validate(model[1], x, y, cv = 5, return_train_score=True, scoring="r2")
    print(model[0])
    train_acc = round((result["train_score"].mean())*100, 2)
    test_acc = round((result["test_score"].mean())*100, 2)
    print(f"Train R2 score : {train_acc}")
    print(f"Test R2 score : {test_acc}")
    print("\n")
```

Random Forest Hyperparameter Tuning Using RandomSearch 

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
parameters = {
    "n_estimators": [100,200,300],
    "max_depth": [None, 5, 10],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "bootstrap": [True, False]
}

# tunning using random search 
random_search = RandomizedSearchCV(
    param_distributions=parameters, # specify the parameter distribution to sample from
    estimator=RandomForestRegressor(), # Specify the estimatore to use
    cv=5, # number of cross validation folds
    return_train_score=True, # Incude train score in the results
    scoring="r2", # Scoring metric to optimize
    n_jobs = -1 # Use all available CPU cores for computation 
)
```

```python
random_search.fit(x, y)
```

```python
random_search.best_params_
```

```python
random_search.cv_results_["mean_train_score"].mean()*100
```

```python
random_search.cv_results_["mean_test_score"].mean()*100
```

```python
## as you can see above we have reduced the difference hence it is a good model 
```

```python
RF = RandomForestRegressor(n_estimators=100,
                           min_samples_split=5,
                           min_samples_leaf=2,
                           max_depth=5,
                           bootstrap=True)
```

```python
RF.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
predicted_train_data = RF.predict(x_train)
train_score = r2_score(y_true=y_train, y_pred=predicted_train_data)
predicted_test_data = RF.predict(X_test)
test_score = r2_score(y_true=y_test, y_pred=predicted_test_data)
print(f"R-squared for training data using random forest: {round(train_score,2)*100}%")
print(f"R-squared for testing data using random forest : {round(test_score,2)*100}%")
```
